```python
from pathlib import Path
from PIL import Image

# 👉 UPDATE THIS PATH to your actual extracted folder
data_dir = Path(r"C:\Users\AbdulHafeez\Brain_Tumor\extracted")

# 1. List class folders
classes = [p.name for p in data_dir.iterdir() if p.is_dir()]
print("Classes found:", classes)

# 2. Count number of images in each class
counts = {}
for c in classes:
    counts[c] = len(list((data_dir / c).glob("*.*")))
print("\nImage counts per class:")
print(counts)

# 3. Sample image details
print("\nSample image details:")
for c in classes:
    sample = next((data_dir / c).glob("*.*"))
    img = Image.open(sample)
    print(f"{c}: {sample.name}, Size={img.size}, Mode={img.mode}")
```
