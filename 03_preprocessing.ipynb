```python
from pathlib import Path
from PIL import Image
import os

# Paths
raw_dir = Path(r"C:\Users\AbdulHafeez\Brain_Tumor\extracted")
prep_dir = Path(r"C:\Users\AbdulHafeez\Brain_Tumor\preprocessed")

IMG_SIZE = (224, 224)

# Loop through all images and preprocess
for class_name in [p.name for p in raw_dir.iterdir() if p.is_dir()]:
    print(f"Processing class: {class_name}")

    # Create class folder inside preprocessed (if not exists)
    (prep_dir / class_name).mkdir(parents=True, exist_ok=True)

    for img_path in (raw_dir / class_name).glob("*.*"):
        try:
            img = Image.open(img_path)

            # Convert RGBA -> RGB
            if img.mode != "RGB":
                img = img.convert("RGB")

            # Resize image
            img = img.resize(IMG_SIZE)

            # Save to preprocessed folder
            save_path = prep_dir / class_name / img_path.name
            img.save(save_path)

        except Exception as e:
            print("Error processing:", img_path, e)

print("Preprocessing Completed!")
```
