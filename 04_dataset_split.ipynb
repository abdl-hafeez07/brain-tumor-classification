```python
from pathlib import Path
import shutil
import random
import os

# Paths
prep_dir = Path(r"C:\Users\AbdulHafeez\Brain_Tumor\preprocessed")
split_dir = Path(r"C:\Users\AbdulHafeez\Brain_Tumor\dataset_split")

# Clean the split directory first (IMPORTANT)
if split_dir.exists():
    shutil.rmtree(split_dir)  # delete entire split folder
split_dir.mkdir()  # recreate empty folder

# Create fresh train/val/test folders
for folder in ["train", "val", "test"]:
    for class_name in [c.name for c in prep_dir.iterdir() if c.is_dir()]:
        (split_dir / folder / class_name).mkdir(parents=True, exist_ok=True)

# Split ratios
train_ratio = 0.7
val_ratio = 0.15
test_ratio = 0.15

for class_name in [c.name for c in prep_dir.iterdir() if c.is_dir()]:
    images = list((prep_dir / class_name).glob("*.*"))
    random.shuffle(images)

    total = len(images)
    train_end = int(total * train_ratio)
    val_end = train_end + int(total * val_ratio)

    train_images = images[:train_end]
    val_images = images[train_end:val_end]
    test_images = images[val_end:]

    for img in train_images:
        shutil.copy(img, split_dir / "train" / class_name / img.name)

    for img in val_images:
        shutil.copy(img, split_dir / "val" / class_name / img.name)

    for img in test_images:
        shutil.copy(img, split_dir / "test" / class_name / img.name)

    print(f"{class_name}: Train={len(train_images)}, Val={len(val_images)}, Test={len(test_images)}")

print("\nDataset split completed successfully!")
```
