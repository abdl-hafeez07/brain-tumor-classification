```python
import os
import shutil
import random

SOURCE_DIR = "/mnt/c/Users/AbdulHafeez/Brain_Tumor/preprocessed"
DEST_DIR   = "/mnt/c/Users/AbdulHafeez/Brain_Tumor/Data_Split"

CLASSES = ["glioma", "meningioma", "no_tumor", "pituitary"]

TRAIN_RATIO = 0.70
VAL_RATIO   = 0.15
TEST_RATIO  = 0.15

SEED = 229
random.seed(SEED)
```

```python
for split in ["train", "val", "test"]:
    for cls in CLASSES:
        os.makedirs(os.path.join(DEST_DIR, split, cls), exist_ok=True)
```

```python
total_counts = {"train": 0, "val": 0, "test": 0}

print("\n📊 CLASS-WISE SPLIT (70 / 15 / 15)\n")

for cls in CLASSES:
    cls_path = os.path.join(SOURCE_DIR, cls)
    images = [f for f in os.listdir(cls_path)
              if f.lower().endswith((".png", ".jpg", ".jpeg"))]

    random.shuffle(images)
    n = len(images)

    train_n = round(n * TRAIN_RATIO)
    val_n   = round(n * VAL_RATIO)
    test_n  = n - train_n - val_n   # guarantees total consistency

    train_imgs = images[:train_n]
    val_imgs   = images[train_n:train_n + val_n]
    test_imgs  = images[train_n + val_n:]

    for img in train_imgs:
        shutil.copy(os.path.join(cls_path, img),
                    os.path.join(DEST_DIR, "train", cls, img))

    for img in val_imgs:
        shutil.copy(os.path.join(cls_path, img),
                    os.path.join(DEST_DIR, "val", cls, img))

    for img in test_imgs:
        shutil.copy(os.path.join(cls_path, img),
                    os.path.join(DEST_DIR, "test", cls, img))

    total_counts["train"] += len(train_imgs)
    total_counts["val"]   += len(val_imgs)
    total_counts["test"]  += len(test_imgs)

    print(f"{cls}: train={len(train_imgs)}, val={len(val_imgs)}, test={len(test_imgs)}")
```

```python
print("\n✅ FINAL SPLIT COUNTS")
print(f"Training:   {total_counts['train']}")
print(f"Validation: {total_counts['val']}")
print(f"Test:       {total_counts['test']}")
print(f"Total:      {sum(total_counts.values())}")
```
